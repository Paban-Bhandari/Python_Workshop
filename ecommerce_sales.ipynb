```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv('data/ecommerce_sales_analysis.csv')
print(df)
```

```python
df['total_sale'] = df[[f'sales_month_{i}' for i in range(1, 13)]].sum(axis=1)
print(df.head())
```

```python
# Group by category and sum the total sales
category_sales = df.groupby('category')['total_sale'].sum().reset_index()

# Print to check if data is correctly aggregated
print(category_sales)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(category_sales['category'], category_sales['total_sale'])
plt.title('Total sales by category')
plt.xlabel('Category')
plt.ylabel('Total sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Average Review score by category:

category_review_score = df.groupby('category')['review_score'].mean().reset_index()
print(category_review_score)
plt.figure(figsize=(10, 6))
plt.bar(category_review_score['category'], category_review_score['review_score'])
plt.title('Average review score by category')
plt.xlabel('category')
plt.ylabel('Average Review score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assuming df is your DataFrame
top_10_products = df.nlargest(10, 'total_sale')  # Using 'total_sale' instead of 'total_sales'
print(top_10_products)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top_10_products['product_name'], top_10_products['total_sale'])
plt.title('Top 10 products by sales')
plt.xlabel('Product Name')
plt.ylabel('Total sales')
plt.xticks(rotation=45)  # Corrected rotation syntax
plt.tight_layout()
plt.show()
```

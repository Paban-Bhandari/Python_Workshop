```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/Top Youtubers Dataset.csv', encoding='ISO-8859-1')
print(df)
```

```python
#total engagement
df['total_engagement'] = df[['Subscribers', 'Video Views', 'Video Count']].sum(axis=1)
print(df.head())
```

```python
# Group by Category and sum the Subscribers
category_subscribers = df.groupby('Category')['Subscribers'].sum().reset_index()
print(category_subscribers)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(category_subscribers['Category'], category_subscribers['Subscribers'], color='skyblue')
plt.title('Total Subscribers by Category')
plt.xlabel('Category')
plt.ylabel('Total Subscribers (in billions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Top 10 YouTubers by Subscribers
top_10_youtubers = df.nlargest(10, 'Subscribers')
print(top_10_youtubers)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_10_youtubers['Youtuber'], top_10_youtubers['Subscribers'], color='skyblue')
plt.title('Top 10 YouTubers by Subscribers')
plt.xlabel('YouTuber')
plt.ylabel('Subscribers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Average Video Count by Category
category_avg_video_count = df.groupby('Category')['Video Count'].mean().reset_index()
print(category_avg_video_count)

plt.figure(figsize=(10, 6))
plt.bar(category_avg_video_count['Category'], category_avg_video_count['Video Count'], color='lightgreen')
plt.title('Average Video Count by Category')
plt.xlabel('Category')
plt.ylabel('Average Video Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
category_video_count = df.groupby('Category')['Video Count'].sum().reset_index()
total_video_count = category_video_count['Video Count'].sum()
category_video_count['Percentage'] = (category_video_count['Video Count'] / total_video_count) * 100
category_video_count = category_video_count.sort_values(by='Percentage', ascending=False)
print(category_video_count)

# Pie Chart
plt.figure(figsize=(10, 10))
plt.pie(
    category_video_count['Video Count'],
    labels=category_video_count['Category'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Category Distribution by Video Count')
plt.tight_layout()
plt.show()
```
